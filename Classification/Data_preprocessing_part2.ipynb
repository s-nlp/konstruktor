## Before training

```python
import json
data_train_all = json.loads(open('data_all_FALL/mintaka_train_dev_simple.json').read())
```

```python
len(data_train_all)
```

```python
for i in range(len(data_train_all)):
    new_elems = data_train_all[i]['labels'][0].split(', ')
    data_train_all[i]['labels'] = new_elems
```

```python
data_train_all
```

```python
# data_train_all = json.loads(open('data/training_data.json').read())
# data_train_all = json.loads(open('data/training_data_filt.json').read())
labels = []
for elem in data_train_all:
    labels.append(', '.join(elem['labels']))
    
from sklearn.model_selection import train_test_split
data_train, data_val = train_test_split(data_train_all, random_state=123, test_size=0.1, stratify = list(labels))
# data_train, data_val = train_test_split(data_train_all, random_state=123, test_size=0.1)

print(len(data_train), len(data_val))

from collections import Counter
term2count = Counter([x for docu in data_train for x in docu['labels']])
```

```python
len(data_train_all)*0.1
```

```python
import json
data_test = json.loads(open('data_all_FALL/mintaka_dates').read())
```

```python
FREQ_CUTOFF = 0 
int_list = [term for term, count in term2count.items() if count>=FREQ_CUTOFF]
int_list = [str(elem) for elem in int_list]
term_freq = sorted(int_list)
# term_freq = sorted([term for term, count in term2count.items() if count>=FREQ_CUTOFF])
labels_ref = sorted([z for z in set([y for x in data_train for y in x['labels']]) if z in term_freq]) 
print(len(term2count), len(labels_ref))
class_freq = [term2count[x] for x in labels_ref]
train_num = len(data_train)
```

```python
import pickle
pickle.dump(data_train, open('data_all_FALL/data_train.rand123','wb')) 
pickle.dump(data_val, open('data_all_FALL/data_val.rand123','wb')) 
# pickle.dump(data_test, open('data/data_test.rand123','wb')) 
pickle.dump(labels_ref, open('data_all_FALL/labels_ref.rand123','wb')) 
pickle.dump(class_freq, open('data_all_FALL/class_freq.rand123','wb')) 
pickle.dump(train_num, open('data_all_FALL/train_num.rand123','wb')) 
```
